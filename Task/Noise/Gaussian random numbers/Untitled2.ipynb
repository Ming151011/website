```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rand():
    return np.random.normal(0,1)
```

```python
def Dnsqt(wla,T):
    hdk=7.63823258e-12
    return np.sqrt(6*hdk**2*wla/(np.exp(hdk*wla/T)-1)/T**2/np.pi)
```

```python
M=2**21
tsim=7e-12
eps0 = 8.854187e-12
hbar = 1.05457182e-34
c = 2.99792458e8
kb = 1.380649e-23
deltap = np.sqrt(1/3/eps0/hbar/c)*kb
def Es(tj,T):
    hdk=7.63823258e-12
    Es0 = rand()*np.sqrt(6*hdk/np.pi/T)
    Esm = 0
    for j in range(1,M-1,1):
        wj = 2*np.pi*j/tsim
        Esm = Esm + (rand()*np.cos(wj*tj)-rand()*np.sin(wj*tj))*Dnsqt(wj,T)
    wf = 2*np.pi*M/tsim
    Esf =  (rand()*np.cos(wf*tj)-rand()*np.sin(wf*tj))*Dnsqt(wf,T)
    return deltap*T*(Es0 + 2*Esm + Esf)/np.sqrt(tsim)
```

```python
Dnsqt(1e12,30000)
```

```python
tl = np.linspace(0,2e-15,100)
T = 50000
plt.plot(tl,Es(tl,T))
```
