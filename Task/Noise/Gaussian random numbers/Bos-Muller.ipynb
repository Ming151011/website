```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as spy
import math as m 
```

```python
#產生隨機亂數產生器
rng = np.random.RandomState(seed=1)

#從seed=1的亂數產生器生成兩個方向1000個亂數
u1, u2 = rng.uniform(size=100000), rng.uniform(size=100000)

# 2. Tranform U1 to s
s = -np.log(u1)

# 3. Transform U2 to theta
thetas = 2*np.pi*u2

# 4. Convert s to r
r = np.sqrt(2*s)

# 5. Calculate x and y from r and theta
Ml, Nl = r*np.cos(thetas), r*np.sin(thetas)
```

```python
print(Ml, Nl)
```

```python
plt.plot(Ml, Nl)
```
