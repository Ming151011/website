```python
import meep as mp
import math as mt
import matplotlib.pyplot as plt
import numpy as np
```

```python
xlab = ['Γ','X','M','Γ']
#r = [1,1,2**0.5]
Γ=0
X=1
M=1+1
Γ2=1+1+2**0.5
rxlab = [Γ,X,M,Γ2]

a1 = np.linspace(0,1/3,1000)
a2 = 1/3
b = (a1**2+a2**2)**0.5
omega = b 
x = np.linspace(rxlab[1],rxlab[2],1000)

plt.plot([rxlab[0],rxlab[1]],[0,1/3])
plt.plot(x,omega)
plt.plot([rxlab[2],rxlab[3]],[2**0.5/3,0])

plt.xticks(rxlab,labels=xlab)
plt.ylabel('Frequency(c/a)')
plt.text(rxlab[1],omega[0] ,'0.333',fontsize=14)
plt.text(rxlab[2],omega[-1],'0.471',fontsize=14)
plt.grid()
```
