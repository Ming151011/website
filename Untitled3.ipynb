```python
import meep as mp
import math as mt
import matplotlib.pyplot as plt
import numpy as np
```

```python
xlab = ['Γ','M','K','Γ']
#r = [3**0.5,1,2]
Γ=0
M=3**0.5
K=1+3**0.5
Γ2=1+2+3**0.5
rxlab = [Γ,M,K,Γ2]

a1 = np.linspace(0,1/3,1000)
a2 = 1/3**0.5
b = (a1**2+a2**2)**0.5
omega = b
x = np.linspace(rxlab[1],rxlab[2],1000)

plt.plot([rxlab[0],rxlab[1]],[0,1/(3**0.5)])
plt.plot(x,omega)
plt.plot([rxlab[2],rxlab[3]],[2/3,0])

plt.xticks(rxlab,labels=xlab)
plt.ylabel('Frequency(c/a)')
plt.text(rxlab[1],omega[0] ,'0.557',fontsize=14)
plt.text(rxlab[2],omega[-1],'0.667',fontsize=14)
plt.grid()
```
