```python
 import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
d = 1.0  # Position from the lens to the object
n_values = [10, 1.5, 1.05, 0.95, 0.01, -10]  # List of refractive index values

# Create an array of x/d values
x_values = np.linspace(-2*d, 2*d, 10000)

# Plot the curves
for n in n_values:
    z_values = d * (((x_values/d)**2 / ((n-1)/(n+1))) + 1)**0.5 + (n/(n+1)) * d
    plt.plot(z_values/d, x_values/d, label=f'n={n}')

# Set plot attributes
plt.xlabel('z/d')
plt.ylabel('x/d')
plt.xlim(0,10)
plt.legend()
plt.grid(True)
plt.savefig('Refractive Index1.png')#, transparent=True)
# Display the plot
plt.show()
```
